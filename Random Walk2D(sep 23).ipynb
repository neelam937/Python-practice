## Random Walk-2D from CLASS

```python
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def move2d(xi,yi):
    direction= random.choice(['x','y'])
    if direction == 'x':
        r=random.choice([-1,1])
        if r==1: xf=xi+1
        else: xf= xi-1
        yf=yi
    if direction == 'y':
        r=random.choice([-1,1])
        if r==1: yf=yi+1
        else: yf=yi-1
        xf=xi
    return xf,yf
def Walk2D(N,pos):
    X=[];Y=[]
    x0=pos[0];y0=pos[1]
    k=0
    while k<N:
        x,y=move2d(x0,y0)  #x,y means xf,yf
        X.append(x)
        Y.append(y)
        x0=x
        y0=y
        k=k+1
    return X,Y
```

```python
N=8000
U=Walk2D(N,[0,0])
V=Walk2D(N,[0,0])
O=Walk2D(N,[0,0])
P=Walk2D(N,[0,0])
Q=Walk2D(N,[0,0])
R=Walk2D(N,[0,0])
#print('U:',U)
#print('V:',V)
#print('O:',O)
#print('P:',P)
#print('Q:',Q)
#print('R:',R)
    
```

```python
plt.figure(figsize=[10,10])
plt.plot(U[0],U[1],'o-')
plt.plot(V[0],V[1],'o-')
plt.plot(O[0],O[1],'o-')
plt.plot(P[0],P[1],'o-')
plt.plot(Q[0],Q[1],'o-')
plt.plot(R[0],R[1],'o-')
plt.grid(True)
plt.xlabel('X',fontsize=14,fontweight='bold')
plt.ylabel('Y',fontsize=14,fontweight='bold')
plt.title('Browninan Motion in 2D')
plt.savefig('Random2D.png')
```

## Random Walk-2D from LOOP

```python
import numpy as np
import matplotlib.pyplot as plt
import random as random
import seaborn as sns
```

```python
class walk2D(object):
    def __init__(self,x_initial,y_initial,step,number):
        self.x0=x_initial
        self.y0=y_initial
        self.d=step
        self.N=number
    def walk2D(self,xi,yi):
        direction=random.choice(['x','y'])
        if direction=='x':
            r=random.choice([1,-1])
            xi=xi+r
        if direction=='y':
            r=random.choice([1,-1])
            yi=yi+r
        return xi,yi
    def walking2d(self):
        ##from where to start##
        X=[]
        Y=[]
        x0=self.x0
        y0=self.y0
        X.append(self.x0)
        Y.append(self.y0)
        k=0
        while k<self.N:
            x,y=self.walk2D(x0,y0) ##move a step
            X.append(x)  ## collecting x 
            Y.append(y)  ##collecting y
            x0=x ## set previous position for next step
            y0=y
            k=k+1
        return X,Y
```

```python
#type-1
W=walk2D(0,0,1,1000)
X,Y=W.walking2d()
len(X) #number of positions
#len(Y)
```

```python
plt.figure(figsize=[10,10])
for a in range(100):
    X,Y=W.walking2d()
    plt.plot(X,Y)
plt.xlabel('x-axis',fontsize=14,fontstyle='italic',color='black')
plt.ylabel('y-axis',fontsize=14,fontstyle='italic',color='black')
plt.grid(True)
plt.title('Walk-2D',fontsize=20,fontstyle='oblique',color='b')
plt.show()
```
