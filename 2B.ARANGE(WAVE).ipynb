# Wave Propagation


A wave propagation is defined as
$ Y= A\sin(\omega t+kx)$

Differential equation
$\large{\frac{\partial^{2}U}{dt^{2}}=c^{2}\frac{\partial^{2}U}{dx^{2}}}$

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
````

Define amplitude,angular frequency and wave factor

````python
amp=1.0
freq=1
w=2*np.pi*freq #angular frequency
k=2*np.pi #wave factor
````

Calculate frequency,wavelength and velocity

````python
wavelength=2*np.pi/k
velocity=freq*wavelength
(freq,wavelength,velocity)
````

### Plot the different configuration of wave motion

````python
xs=np.arange(0,5,1/20)
ts=np.arange(0,5,1/20)
````

````python
len(xs),len(ts)
````

### 1.Experiment:Plot the particle displacement keeping time fixed

When we fix the time,it is the photograph of the wave motion at specific time t0.It can be represented as 
$Y= A\ sin(kx+ \phi_{0})$

A special case is **closed vibrating string**

````python
plt.figure(figsize=[18,4])
t0=0.0
y=amp*np.sin(w*t0 -k*xs)
plt.stem(y,use_line_collection=True)
plt.xlabel("Distance")
plt.ylabel("Amplitude")
plt.title("Particle displacement at t=" +str(t0))
````

````python
plt.figure(figsize=[18,18])
t0s= [0.0,0.1,0.2,0.3,0.4]
for i,t0 in enumerate(t0s):
    plt.subplot(len(t0s),1,1+i)
    y+amp*np.sin(w*t0 -k*xs)
    plt.stem(y,use_line_collection=True)
    plt.title('Particle displacement at t='+ str(t0))
plt.show()
````

### 2.Experiment:Plot the particle displacement keeping position fixed

When we fix the position and look at a single particle displacement, it behaves like a harmonic oscillator.It can be represented as 

$Y=A\sin(\omega t+ \phi_{0})$

A special case is a **harmonic oscillator**

````python
plt.figure(figsize=[18,4])
x0=0.0
y=amp*np.sin(w*ts - k*x0)
plt.stem(y,use_line_collection=True)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Particle motion at x="+ str(x0))
````

````python
plt.figure(figsize=[18,18])
x0s=[0.0,0.1,0.2,0.3,0.4]
for i,x0 in enumerate(t0s):
    plt.subplot(len(x0s),1,1+i)
    y=amp*np.sin(w*ts - k*x0)
    plt.stem(y,use_line_collection=True)
    plt.title("Particle displacement at x=" +str(x0))
plt.show()
````
