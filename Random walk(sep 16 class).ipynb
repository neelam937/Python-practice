# RANDOM WALK

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import random
```

## One dimensional random walk

```python
X=[] #empty list to collect position of walk
T=[] #empty list to collect time of walk
x= 0 #starting point origin
d= 1 #step length
N= 1000 #number of steps
for t in range(N): #using for loop
    x = x + d*random.choice([-1,1])
    T.append(t)
    X.append(x)
print(X)
print(T)
```

```python
plt.figure(figsize=[10,4])
plt.plot(T,X,marker='o',color='b',lw=2)
plt.xlabel('Time',fontsize=10)
plt.ylabel('Position',fontsize=10)
plt.xticks(fontsize=12,color='g')
plt.yticks(fontsize=12,color='g')
```

## One dimensonal Random Walk using function

```python
def walk1d(x,N,d):
    X=[]
    T=[]
    for t in range(N):
        x=x+d*random.choice([-1,1])
        T.append(t)
        X.append(x)
    return X,T
X1,T1=walk1d(x=0,N=1000,d=1) #for particle first
X2,T2=walk1d(x=0,N=1000,d=1) #for particle second
```

```python
plt.figure(figsize=[10,5])
plt.plot(T1,X1,'o','g',lw=1)
plt.plot(T2,X2,'-','r',lw=1)
plt.xlabel('Time',fontsize=10)
plt.ylabel('Position',fontsize=10)
plt.xticks(fontsize=14,color='g')
plt.yticks(fontsize=14,color='g')
         
```

## 1D Random walk using class

```python
class Walk1D():
    def __init__(self,N,d,x0):
        self.N=N
        self.X=[] #empty list to collect position
        self.d=d
        self.x0=x0
    def walk1d(self):
        x=self.x0
        k=0
        while k<N:
            x=x+d*random.choice([-1,1])
            self.X.append(x)
            k=k+1
        return self.X   
```

```python
X1=Walk1D(N=1000,d=1,x0=0).walk1d()
X2=Walk1D(N=1000,d=1,x0=0).walk1d()
X3=Walk1D(N=1000,d=1,x0=0).walk1d()
X4=Walk1D(N=1000,d=1,x0=0).walk1d()
T= [i for i in range(len(X1))]
print(len(X1))
```

```python
plt.figure(figsize=[12,5])
plt.plot(T,X1,'-','r')
plt.plot(T,X2,'-','g')
plt.plot(T,X3,'-','b')
plt.plot(T,X4,'-','m')
```
